```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from cycler import cycler
import matplotlib.cm as cm
# Set Dark2 as the default color cycle
dark2 = cm.get_cmap('Dark2').colors
plt.rcParams['axes.prop_cycle'] = cycler(color=dark2)
%matplotlib widget
from matplotlib.ticker import ScalarFormatter
```

```python
def CalcMassCylinder(det_size, thick):
    rho_Cu   = 8960 # kg/m^3x
    return np.pi*0.25*rho_Cu * ( (det_size+2*thick)**3 - det_size**3 )

Detsize = np.array([6.182, 3.615, 2.870, 2.507, 2.114])
MCu_12cm = CalcMassCylinder(Detsize, 12e-2)

pressures = [1, 5, 10, 15, 25]

# Enriched
Enr_Win_nubb_12cm = np.array([5837, 7830, 8330, 8617, 8919])/10000
Enr_Win_single_12cm = np.array([5827, 8232, 8892, 9393, 9747])/100000
Enr_Win_Bi_12cm = np.array([708, 3241, 6891, 11105, 19419])/100000000
Enr_Win_Tl_12cm = np.array([3153, 13391, 22002, 28604, 37958])/10000000 

Nat_Win_nubb_12cm = np.array([7330, 8562, 8985, 9253, 9441])/10000
Nat_Win_single_12cm = np.array([7703, 9251, 9532, 9783, 10146])/100000
Nat_Win_Bi_12cm = np.array([19962, 112142, 227622, 322638, 493132])/1000000000
Nat_Win_Tl_12cm = np.array([9562, 29985, 42677, 49219, 57306])/10000000 

fig = plt.figure(figsize=(10,10));

plt.subplot(221);
plt.plot(pressures, Enr_Win_nubb_12cm, linewidth = 2, color = "k", label = "Enriched")
plt.plot(pressures, Nat_Win_nubb_12cm, linewidth = 2, color = "k", linestyle="--", label = "Natural")
plt.xlabel("Pressure [bar]", fontsize=15)
plt.ylabel(r"Containment Efficiency ($\eta^{\mathrm{cont}}$)", fontsize=15)
plt.tick_params(axis='both', which='both', labelsize=15)
plt.title(r"$0 \nu \beta \beta$ (2.3 - 2.6 MeV)", fontsize = 15)
plt.legend()
plt.grid()
plt.tight_layout()


plt.subplot(222);
plt.plot(pressures, Enr_Win_single_12cm, linewidth = 2, color = "k", label = "Enriched")
plt.plot(pressures, Nat_Win_single_12cm, linewidth = 2, color = "k", linestyle="--", label = "Natural")
plt.xlabel("Pressure [bar]", fontsize=15)
plt.ylabel(r"Containment Efficiency ($\eta^{\mathrm{cont}}$)", fontsize=15)
plt.tick_params(axis='both', which='both', labelsize=15)
plt.title("$^{137}$Xe (2.3 - 2.6 MeV)", fontsize = 15)
plt.legend()
plt.grid()
plt.tight_layout()


plt.subplot(223);
plt.plot(pressures, Enr_Win_Bi_12cm, linewidth = 2, color = "k", label = "Enriched")
plt.plot(pressures, Nat_Win_Bi_12cm, linewidth = 2, color = "k", linestyle="--", label = "Natural")
plt.xlabel("Pressure [bar]", fontsize=15)
plt.ylabel(r"Containment Efficiency ($\eta^{\mathrm{cont}}$)", fontsize=15)
plt.tick_params(axis='both', which='both', labelsize=15)
plt.title("$^{214}$Bi (2.3 - 2.6 MeV)", fontsize = 15)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.legend()
plt.grid()
plt.tight_layout()

plt.subplot(224);
plt.plot(pressures, Enr_Win_Tl_12cm, linewidth = 2, color = "k", label = "Enriched")
plt.plot(pressures, Nat_Win_Tl_12cm, linewidth = 2, color = "k", linestyle="--", label = "Natural")
plt.xlabel("Pressure [bar]", fontsize=15)
plt.ylabel(r"Containment Efficiency ($\eta^{\mathrm{cont}}$)", fontsize=15)
plt.tick_params(axis='both', which='both', labelsize=15)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.legend()
plt.title("$^{208}$Tl (2.3 - 2.6 MeV)", fontsize = 15)
plt.grid()
plt.tight_layout()

plt.savefig(f"../../../Projects/Atmospheric TPC/PaperDraft/ContainmentRates.pdf")
```
